```python
import numpy as np
import math
import matplotlib.pyplot as plt

# plt.style.use("ggplot")

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
```

```python
x = np.linspace(-2, 2, 150)
y = np.linspace(-2, 2, 150)
X, Y = np.meshgrid(x, y)

def plot_surface(ax, Z, title, elev=35, azim=-20, zlabel=r"$f(x_1,x_2)$"):
    surf = ax.plot_surface(X, Y, Z, linewidth=0, antialiased=True, alpha=0.8)
    Z0 = Z[(np.abs(y-0)).argmin(), (np.abs(x-0)).argmin()]
    ax.scatter([0], [0], [Z0], s=80, color="black", marker="o")
    ax.set_title(title, fontsize=10)
    ax.set_xlabel(r"$x_1$")
    ax.set_ylabel(r"$x_2$")
    ax.set_zlabel(zlabel)
    ax.view_init(elev=elev, azim=azim)
    try:
        ax.set_box_aspect((1, 1, 0.7))
    except Exception:
        pass

fig = plt.figure(figsize=(15, 5))
axs = [fig.add_subplot(1, 3, i+1, projection="3d") for i in range(3)]

# f(x) = x1^2 + x2^2
Z1 = X**2 + Y**2
plot_surface(axs[0], Z1, r"$x_1^2 + x_2^2$  (strict local min at $(0,0)$)")

# f(x) = x1^2 - x2^2
Z2 = X**2 - Y**2
plot_surface(axs[1], Z2, r"$x_1^2 - x_2^2$  (saddle at $(0,0)$)")

# f(x) = x1^4 + x2^4
Z3 = X**4 + Y**4
plot_surface(axs[2], Z3, r"$x_1^4 + x_2^4$  (SOSC inconclusive at $(0,0)$)")

plt.tight_layout()
plt.savefig("../../../figs/optimization/critical_points.pdf", bbox_inches="tight", transparent=True)
plt.show()
```

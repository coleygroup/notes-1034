```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
```

```python
x = np.linspace(-1, 5.0, 10000)

f = (x - 2.0) ** 2

mus = [10.0, 1.0]

plt.figure(figsize=(6, 4))
plt.plot(x, f, label=r"$f(x)=(x-2)^2$")

for mu in mus:
    phi = f - mu * np.log(x)  # logarithmic barrier: -mu * log(h(x)) with h(x)=x
    plt.plot(x, phi, label=rf"$\phi(x,\mu={mu})$")

plt.axvline(0.0, linestyle=":", linewidth=1.0, color="black")
plt.text(-0.16, 6.5, r"Constraint $x>0$", rotation=90, va="bottom")

plt.xlabel("$x$")
plt.ylabel("objective value")
# plt.title("Logarithmic Barrier for $x>0$")
plt.legend()
plt.grid(True)
plt.tight_layout()

pdf_path = '../../../figs/optimization/barrier.pdf'
plt.savefig(pdf_path, bbox_inches="tight")
plt.show()
```

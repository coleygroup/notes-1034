```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
```

```python
def f(x): return x + x**3
def tangent(x): return x   # because f'(0) = 1
def remainder_bound(x): return x**2

r = 0.7
x = np.linspace(-1, 1, 500)
mask = np.abs(x) <= r

plt.figure(figsize=(6, 4))
plt.plot(x, f(x), color='C0', lw=2, label=r'$f(x)=x+x^3$')
plt.plot(x, tangent(x), color='C1', ls='--', lw=2, label=r'Tangent at $x^*$')
plt.fill_between(x[mask],
                tangent(x[mask]) - remainder_bound(x[mask]),
                tangent(x[mask]) + remainder_bound(x[mask]),
                color='C2', alpha=0.3,
                label=r'$\pm(\Delta x)^2$ band for $\left| x-x^* \right| \le r$'
                )

plt.scatter([0], [0], color='C0', zorder=5)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)

plt.xlabel(r'$x - x^*$')
plt.ylabel(r'$y$')
plt.xlim(-r, r)
plt.ylim(-0.8, 0.8)
plt.legend(loc='upper left', framealpha=1)
plt.tight_layout()
plt.grid(True)

# plt.savefig('../../../figs/nle/tangent_band.pdf', bbox_inches='tight', transparent=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
```

```python
def f1(x, y):
    return x * np.exp(y) - 1

def f2(x, y):
    return x * np.exp(-y) - 1

x = np.linspace(0, 2, 400)
y = np.linspace(-1, 1, 400)
X, Y = np.meshgrid(x, y)

Z1 = f1(X, Y)
Z2 = f2(X, Y)

plt.figure()
cs1 = plt.contour(X, Y, Z1, levels=[0], colors='r')
cs2 = plt.contour(X, Y, Z2, levels=[0], colors='b')


# annotate the lines with label in red and blue
plt.text(0.65, 0.55, r'$f_1(x,y)=0$', color='r')
plt.text(0.65, -0.55, r'$f_2(x,y)=0$', color='b')

plt.plot(1, 0, 'k.', markersize=10)

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# plt.savefig('../../../figs/nle/exponential_example.pdf', bbox_inches='tight', transparent=True)
plt.show()
```

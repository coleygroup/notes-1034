```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Leibniz series approximation
def leibniz_pi(num_terms):
    pi_approx = []
    sum = 0.0
    for n in range(num_terms):
        term = (-1)**n / (2*n + 1)
        sum += term
        pi_approx.append(4 * sum)
    return pi_approx

# Ramanujan series approximation
def ramanujan_pi(num_terms):
    pi_approx = []
    factor = (2 * math.sqrt(2)) / 9801
    sum = 0.0
    for n in range(num_terms):
        numerator = math.factorial(4*n) * (1103 + 26390*n)
        denominator = (math.factorial(n))**4 * 396**(4*n)
        term = numerator / denominator
        sum += term
        pi_approx.append(1 / (factor * sum))
    return pi_approx


# compute approximations
n_terms = 10
pi_leibniz = leibniz_pi(n_terms)
pi_ramanujan = ramanujan_pi(n_terms)

# Compute errors
errors_leibniz = [abs(pi - math.pi) for pi in pi_leibniz]
errors_ramanujan = [abs(pi - math.pi) for pi in pi_ramanujan]

# plot
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(range(1, n_terms + 1), errors_leibniz, 'o-', label='Leibniz Series')
ax.plot(range(1, n_terms + 1), errors_ramanujan, 'o-', label='Ramanujan Series')

ax.set_xlim(0, n_terms)
ax.set_yscale('log') 
ax.set_xlabel('Number of Terms')
ax.set_ylabel('Absolute Error')
ax.set_title(r'Convergence of $\pi$ Approximation Series')
ax.legend()
# plt.savefig('../../../figs/linear-algebra/pi-convergence.pdf', bbox_inches='tight')
plt.show()
```

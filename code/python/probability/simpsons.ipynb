```python
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use('ggplot')
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
```

```python
rng = np.random.default_rng(3)

n_a, n_b = 80, 80
x_a = rng.normal(3.0, 0.9, n_a)
x_b = rng.normal(7.0, 0.9, n_b)

y_a = 8.0 - 0.8 * x_a + rng.normal(0, 0.6, n_a)
y_b = 16.0 - 0.8 * x_b + rng.normal(0, 0.6, n_b)

x = np.concatenate([x_a, x_b])
y = np.concatenate([y_a, y_b])

m_a, b_a = np.polyfit(x_a, y_a, 1)
m_b, b_b = np.polyfit(x_b, y_b, 1)
m_all, b_all = np.polyfit(x, y, 1)

fig, ax = plt.subplots(figsize=(7.5, 5.2))

s1 = ax.scatter(x_a, y_a, label="Group A", color="blue")
s2 = ax.scatter(x_b, y_b, label="Group B", color="red")

xx_a = np.linspace(x_a.min()-0.2, x_a.max()+0.2, 50)
xx_b = np.linspace(x_b.min()-0.2, x_b.max()+0.2, 50)
xx_all = np.linspace(x.min()-0.3, x.max()+0.3, 100)

# regression lines
ax.plot(xx_a, m_a * xx_a + b_a, linewidth=3, alpha=0.9, label=f"A fit (slope {m_a:.2f})", color="blue")
ax.plot(xx_b, m_b * xx_b + b_b, linewidth=3, alpha=0.9, label=f"B fit (slope {m_b:.2f})", color="red")
ax.plot(xx_all, m_all * xx_all + b_all, linewidth=4, alpha=0.9, label=f"Overall fit (slope {m_all:.2f})", color="magenta")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, linestyle="--", linewidth=0.7, alpha=0.6)
ax.legend(loc="upper left")
ax.margins(0.05)

fig.tight_layout()
fig.savefig("../../../figs/probability/simpsons_paradox_scatter.pdf", bbox_inches="tight", transparent=True)
plt.show()
```

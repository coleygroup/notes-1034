```python
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use('ggplot')
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
import math
```

```python
n = 20
p = 0.4

k = np.arange(0, n+1)
pmf = np.array([math.comb(n, ki) * (p**ki) * ((1-p)**(n-ki)) for ki in k])

plt.figure(figsize=(7,4))
markerline, stemlines, baseline = plt.stem(k, pmf, linefmt='b', label=f"$B({n}, {p})$")
plt.setp(baseline, visible=False)
# plt.title(f"Binomial PMF (n={n}, p={p})")
plt.xlabel("$k$")
plt.ylabel(r"$p_{\mathsf{k}}(k)$")
plt.xticks(k[::2])
plt.grid(True, linestyle=":", linewidth=0.7)
plt.ylim(0)
plt.legend()
plt.tight_layout()
plt.savefig("../../../figs/probability/binomial_pmf.pdf", bbox_inches="tight", transparent=True)
plt.show()
```

```python
a, b = 0.5, 1.5

x = np.linspace(a-0.2*(b-a), b+0.2*(b-a), 600)
pdf = np.where((x >= a) & (x <= b), 1.0/(b-a), 0.0)

plt.figure(figsize=(7,4))
plt.plot(x, pdf, color="blue", label=f"$U({a}, {b})$")
# plt.title(f"Uniform PDF U({a}, {b})")
plt.xlabel("$x$")
plt.ylabel(r"$p_{\mathsf{x}}(x)$")
# plt.axvline(a, linestyle="--")
# plt.axvline(b, linestyle="--")
plt.legend(loc="upper right")
plt.grid(True, linestyle=":", linewidth=0.7)
plt.ylim(0)
plt.tight_layout()
plt.savefig("../../../figs/probability/uniform_pdf.pdf", bbox_inches="tight", transparent=True)
plt.show()
```

```python
mu = 0.0
sigma = 1.0

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 800)
pdf = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-0.5*((x - mu)/sigma)**2)

plt.figure(figsize=(7,4))
plt.plot(x, pdf, color="blue", label=f"$N({mu}, {sigma**2})$")
# plt.title(f"Normal PDF N({mu}, {sigma**2})")
plt.xlabel("$x$")
plt.ylabel(r"$p_{\mathsf{x}}(x)$")

plt.axvline(mu, linestyle="--", color="black", label=r"$\mu$")
plt.axvline(mu - sigma, linestyle=":", color="black", label=r"$\mu \pm \sigma$")
plt.axvline(mu + sigma, linestyle=":", color="black")
plt.legend()
plt.grid(True, linestyle=":", linewidth=0.7)
plt.ylim(0)
plt.tight_layout()
plt.savefig("../../../figs/probability/normal_pdf.pdf", bbox_inches="tight", transparent=True)
plt.show()
```

```python
lam = 4.0

k_max = int(max(15, lam + 6*np.sqrt(lam)))
k = np.arange(0, k_max+1)
pmf = np.array([(lam**ki) * math.exp(-lam) / math.factorial(ki) for ki in k])

plt.figure(figsize=(7,4))
markerline, stemlines, baseline = plt.stem(k, pmf, linefmt='b', label=r"$\mathrm{Pois}(\lambda)$")
plt.setp(baseline, visible=False)
# plt.title(f"Poisson PMF (={lam})")
plt.xlabel("k")
plt.ylabel(r"$p_{\mathsf{k}}(k)$")
plt.xticks(k[::2])
plt.grid(True, linestyle=":", linewidth=0.7)
plt.legend()
plt.tight_layout()
plt.savefig("../../../figs/probability/poisson_pmf.pdf", bbox_inches="tight", transparent=True)
plt.show()
```

```python
# Exponential PDF plot (analogous to your normal plot)

lam = 1.0  # rate parameter 

# Domain: [0, ~8/] gives a good view of the right tail
x = np.linspace(0.0, 8.0/lam, 800)
pdf = lam * np.exp(-lam * x)

plt.figure(figsize=(7,4))
plt.plot(x, pdf, color="blue", label=rf"$\mathrm{{Exp}}(\lambda={lam})$")
# plt.title(rf"Exponential PDF Exp($\lambda$={lam})")
plt.xlabel("$x$")
plt.ylabel(r"$p_{\mathsf{x}}(x)$")

mu = 1.0/lam
sigma = 1.0/lam
plt.axvline(mu, linestyle="--", color="black", label=rf"$\mu = {mu:.2f}$")
# plt.axvline(max(0, mu - sigma), linestyle=":", color="black", label=r"$\mu \pm \sigma$")
# plt.axvline(mu + sigma, linestyle=":", color="black")
plt.legend()
plt.grid(True, linestyle=":", linewidth=0.7)
plt.ylim(0)
plt.xlim(0, 8.0/lam)
plt.tight_layout()
plt.savefig("../../../figs/probability/exponential_pdf.pdf", bbox_inches="tight", transparent=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# plt.style.use('ggplot')
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
```

```python
sigma_x = 1.0
sigma_y = 2.0

x = np.linspace(-4*sigma_x, 4*sigma_x, 201)
y = np.linspace(-4*sigma_y, 4*sigma_y, 201)
X, Y = np.meshgrid(x, y)

px = (1/np.sqrt(2*np.pi*sigma_x**2)) * np.exp(-X**2/(2*sigma_x**2))
py = (1/np.sqrt(2*np.pi*sigma_y**2)) * np.exp(-Y**2/(2*sigma_y**2))
Z = px * py

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(
    X, Y, Z,
    cmap='viridis',
    linewidth=0,
    antialiased=True,
    rcount=150, ccount=150,
    shade=True
)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel(r'$p_{\mathsf{x},\mathsf{y}}(x,y)$')
# ax.set_title('Bivariate Density: Independent Gaussians')
fig.colorbar(surf, shrink=0.7, aspect=18, pad=0.08, label=r'$p_{\mathsf{x},\mathsf{y}}(x,y)$')
plt.tight_layout()
fig.savefig("../../../figs/probability/bivariate_gaussian_independent.pdf", bbox_inches="tight", transparent=True)
plt.show()
```

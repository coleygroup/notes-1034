```python
import numpy as np
import matplotlib.pyplot as plt
# plt.style.use('ggplot')
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
})
```

```python
sensitivity = 0.85
specificity = 0.999

p = np.logspace(-5, 0, 400)
ppv = (sensitivity * p) / (sensitivity * p + (1 - specificity) * (1 - p))

plt.figure(figsize=(7,4))
plt.plot(p, ppv, color="blue")
plt.xlabel("Prevalence")
plt.ylabel("PPV = P(C | T+)")
# highlight the 1% point
p0 = 0.01
ppv0 = (sensitivity * p0) / (sensitivity * p0 + (1 - specificity) * (1 - p0))
plt.scatter([p0], [ppv0], color="red")
plt.text(p0 + 0.0025, ppv0 - 0.025, rf"  (1\%, {ppv0*100:.1f}\%)", va="center", color="red")
plt.grid(True, linestyle=":", linewidth=0.7)
plt.tight_layout()
plt.xscale("log")
plt.show()
```
